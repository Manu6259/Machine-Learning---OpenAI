### Since NumSatisfactoryTrades and NumTotalTrades are highly correlated, we can remove NumTotalTrades because if we only look at NumTotalTrades, we might assume that a person with many accounts is financially experienced. But if most of those accounts were not managed well, they are actually high risk. Keeping NumSatisfactoryTrades helps ensure that we're focusing on both experience AND responsible credit use.

### We will also remove MaxDelqEver and keep MaxDelq2PublicRecLast12M because the most recent bad payment is more important than one that happened a long time ago. High recent delinquency = The person is currently struggling, which is more useful for predicting loan risk.

```python
# Dropping NumTotalTrades and MaxDelqEver
df_cleaning = df.drop(["NumTotalTrades", "MaxDelqEver"], axis=1)
```

```python
from sklearn.impute import KNNImputer
# Identify columns with -7 values
columns_with_negative_7 = ['MSinceMostRecentDelq', 'MSinceMostRecentInqexcl7days']

# Replace -7 with NaN for imputation
df_cleaning[columns_with_negative_7] = df_cleaning[columns_with_negative_7].replace(-7, float('nan'))

# Select only relevant numerical features for KNN calculations (to reduce memory usage)
selected_features = ['ExternalRiskEstimate', 'NumInqLast6M', 'PercentTradesNeverDelq', 
                     'NetFractionRevolvingBurden', 'NumRevolvingTradesWBalance'] + columns_with_negative_7

# Create a copy with only the relevant features for imputation
knn_data = df_cleaning[selected_features].copy()

# Initialize KNN Imputer with a lower number of neighbors
knn_imputer = KNNImputer(n_neighbors=3)

# Apply KNN Imputation
knn_data[columns_with_negative_7] = knn_imputer.fit_transform(knn_data[columns_with_negative_7])

# Merge imputed values back into the main dataset
df_cleaning[columns_with_negative_7] = knn_data[columns_with_negative_7]

df_cleaning = df_cleaning[df_cleaning != -9]

# Save cleaned dataset
df_cleaning.to_csv("heloc_cleaned_optimized_knn.csv", index=False)

print("KNN Imputation for -7 values completed successfully! Cleaned dataset saved as 'heloc_cleaned_optimized_knn.csv'.")

# remove all the rows with -9 values


```
